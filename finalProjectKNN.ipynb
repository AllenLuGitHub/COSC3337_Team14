```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
files = ['xaa.dat', 'xab.dat', 'xac.dat', 'xad.dat', 'xae.dat', 'xaf.dat', 'xag.dat', 'xah.dat', 'xai.dat']
attributes = ['Compactness', 'Circularity', 'Dist_Circ', 'Radius_Ratio', 'PR_Axis_Ratio', 'Max_Length_Ratio', 'Scatter_Ratio',
             'Enlongatedness', 'PR_Axis_Rectangularity', 'Max_Length_Rectangularity', 'Major_Scaled_Var', 'Minor_Scaled_Var',
             'Scaled Radius Gyration', 'Major_Skew', 'Minor_Skew', 'Minor_Kurtosis', 'Major_Kurtosis', 'Hollow_Ratio',
             'Vehicle_Type']
car_types = ['van', 'saab', 'bus', 'opel']
```

```python
vehicle_df = pd.DataFrame(columns=attributes)
for file in files:
    f = open(file, 'r')
    for line in f:
        l = line.strip().split(' ')
        vehicle_df.loc[len(vehicle_df.index)] = list(map(lambda i: i if i in car_types else int(i), l))
```

```python
for attribute in [att for att in attributes if att != 'Vehicle_Type']:
    vehicle_df[attribute] = vehicle_df[attribute].astype('int')
```

```python
vehicle_df_copy = vehicle_df.copy(deep=True)
```

```python
sns.countplot(x=vehicle_df_copy['Vehicle_Type'])
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler

# all columns except 'vehicle_type'
X = vehicle_df_copy.drop('Vehicle_Type', axis=1)
y = vehicle_df_copy['Vehicle_Type']

# create our scaler object
scaler = StandardScaler()
# use our scaler object to transform/scale our data and save it into X_scaled
X_scaled = scaler.fit_transform(X)
# reassign X to a new DataFrame using the X_scaled values.
X = pd.DataFrame(data=X_scaled, columns=X.columns)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, stratify=y, random_state=42)
```

```python
from sklearn.neighbors import KNeighborsClassifier

# will append scores here for plotting later
test_scores = []
train_scores = []

# testing k values from 1-14
for i in range(1,15):
    # create a model with k=i
    knn = KNeighborsClassifier(i)
    # train the model
    knn.fit(X_train,y_train)
    
    # append scores. 
    train_scores.append(knn.score(X_train,y_train))
    test_scores.append(knn.score(X_test,y_test))
```

```python
sns.lineplot(x=range(1,15), y=train_scores, marker='*', label='Train Score')
sns.lineplot(x=range(1,15), y=test_scores, marker='o', label='Test Score')
plt.title('K vs. Score')
plt.xlabel('K')
plt.ylabel('Score')
plt.show()
```

```python
knn = KNeighborsClassifier(14)

knn.fit(X_train,y_train)
knn.score(X_test,y_test)
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report

y_pred = knn.predict(X_test)
print(classification_report(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
```
